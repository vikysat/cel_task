IN THIS PROBLEM WE NEED TO FIND THE COMPATABILITY SCORE. THE BEST WAY TO DO SO IS TO SEE HOW MANY OF THE ATTRIBUTES REQUIRED BY THE STARTUP MATCH WITH THE ATTRIBUTES OF THE MENTOR.

```python
import pandas as pd
```

1.IMPORT PANDAS AND READ IN THE DATA

```python
pd.read_csv('startup.csv')
```

```python
pd.read_csv('mentor.csv')
```

DEFINING A FUNCTION WHICH FINDS THE NUMBER OF COMMON ELEMENTS IN TWO LISTS


```python
def common(list1, list2):
  count =0
  for i in list1:
    for j in list2:
      if i == j:
        count+=1
  return count


```

STORING THE CSV FILES IN df1 and df2

```python
df1=pd.read_csv('startup.csv')
df2=pd.read_csv('mentor.csv')
```

*DEFINING A FUNCTION WHICH CONVERTS A ROW IN THE DATASET INTO A LIST
* WITHIN THE ROW, WHENEVER THERE IS A QUESTION MARK THE DATA IS SPLIT, I.E THEY ARE STORED AS SEPARATE ELEMENTS
*Nan values are also removed

```python
def make(s):
  result_array = []
  for item in s:
    if pd.notna(item):
       if "?" in str(item):

        # If question mark is present, split the string by the question mark and store each part as a separate element
        split_items = str(item).split("?")
        result_array.extend(split_items)

       else:
        result_array.append(item)
  return result_array





```

LOOP THROUGH ALL THE ROWS IN BOTH THE STARTUP AND MENTOR DATA SET AND CONVERT EACH ROW INTO A LIST

```python
i = 0
s_final = []
m_final =[]
while i<=9:
  s = df1.iloc[i].tolist()
  m = df2.iloc[i].tolist()
  s_final.append(make(s))
  m_final.append(make(m))
  i+=1



```



* USING THE COMMON FUNCTION DEFINED EARLIER , WE FIND THE TOTAL NUMBER OF COMMON ELEMENTS BETWEEN A ROW IN THE STARTUP DATA SET AND A ROW IN THE MENTOR DATA SET.
* THE TOTAL NUMBER OF COMMMON ELEMENTS IS THE COMPATABILITY SCORE
* FOR EACH POSSIBILE COMBINATION OF A STARTUP AND A MENTOR, THERE IS A COMPATABILITY SCORE
* THE COMPATABLITY SCORE FOR EACH POSSIBILITY IS STORED IN A LIST.



```python
compatability_score = []
for startup in s_final:
  for mentor in m_final:
    compatability_score.append(common(startup,mentor))

```
